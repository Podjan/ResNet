# Descriptive Statistic

#### Mengubah tipe data yang kurang sesuai agar memudahkan pengolahan data

```python
# Mengubah data yang bertipe string ke float
dfc['payment'] = dfc['payment'].str.replace('$', '').str.replace(',', '').astype(float)
dfl['price'] = dfl['price'].str.replace('$', '').str.replace(',', '').astype(float)
dfl['weekly_price'] = dfl['weekly_price'].str.replace('$', '').str.replace(',', '').astype(float)
dfl['monthly_price'] = dfl['monthly_price'].str.replace('$', '').str.replace(',', '').astype(float)
dfl['host_response_rate'] = dfl['host_response_rate'].str.replace('%', '').astype(float)
dfl['host_acceptance_rate'] = dfl['host_acceptance_rate'].str.replace('%', '').astype(float)
```

```python
#Mengubah data yang beripe integer ke string
dfl['host_id'] = dfl['host_id'].astype(str)
dfl['latitude'] = dfl['latitude'].astype(str)
dfl['longitude'] = dfl['longitude'].astype(str)
```

#### Marge data listing_id dan calendar

```python
# Menggabungkan DataFrame calendar dan listing berdasarkan 'listing_id'
df_merged = pd.merge(dfc, dfl, on='listing_id', how='left')
df_merged
```

```python
# Definisi fungsi untuk memperbarui status berdasarkan nilai Payment, Day Price, Weekly Price, Monthly Price

def update_status(df):
    """Fungsi untuk memperbarui kolom Day Price Status, Weekly Price Status, dan Monthly Price Status berdasarkan kolom Payment."""
    df['day_price_status'] = df.apply(lambda row: 'Yes' if row['payment'] == row['price'] else 'No', axis=1)
    df['weekly_price_status'] = df.apply(lambda row: 'Yes' if row['payment'] == row['weekly_price'] else 'No', axis=1)
    df['monthly_price_status'] = df.apply(lambda row: 'Yes' if row['payment'] == row['monthly_price'] else 'No', axis=1)
    return df

# Menggunakan fungsi update_status untuk memperbarui status pada DataFrame yang ada
df_result_updated = update_status(df_merged)
df_result_updated[['listing_id', 'date', 'payment', 'day_price_status', 'weekly_price_status', 'monthly_price_status']]
```

**pakai update_status setiap kali ada perubahan di kolom payment**

```python
df_result_updated.info()
```

```python
# Menyaring baris dengan kondisi day_price_status, weekly_price_status, monthly_price_status sama-sama "No"
filtered_df = df_result_updated[(df_result_updated['day_price_status'] == 'No') & 
                                (df_result_updated['weekly_price_status'] == 'No') &
                                (df_result_updated['monthly_price_status'] == 'No')]

# Menghitung jumlah unik listing_id dalam hasil filter
unique_listing_count = filtered_df['listing_id'].nunique()

unique_listing_count
```
