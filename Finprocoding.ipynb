```python
dfc = dfc.rename({'price': 'payment'}, axis=1)
dfc.info()
```

```python
# Menggabungkan DataFrame calendar dan listing berdasarkan 'listing_id'
df_merged = pd.merge(dfc, dfl, on='listing_id', how='left')
df_merged
```

```python
# Definisi fungsi untuk memperbarui status berdasarkan nilai Payment, Day Price, Weekly Price, Monthly Price

def update_status(df):
    """Fungsi untuk memperbarui kolom Day Price Status, Weekly Price Status, dan Monthly Price Status berdasarkan kolom Payment."""
    df['day_price_status'] = df.apply(lambda row: 'Yes' if row['payment'] == row['price'] else 'No', axis=1)
    df['weekly_price_status'] = df.apply(lambda row: 'Yes' if row['payment'] == row['weekly_price'] else 'No', axis=1)
    df['monthly_price_status'] = df.apply(lambda row: 'Yes' if row['payment'] == row['monthly_price'] else 'No', axis=1)
    return df

# Menggunakan fungsi update_status untuk memperbarui status pada DataFrame yang ada
df_result_updated = update_status(df_merged)
df_result_updated[['listing_id', 'date', 'payment', 'day_price_status', 'weekly_price_status', 'monthly_price_status']]
```

**pakai update_status setiap kali ada perubahan di kolom payment**
